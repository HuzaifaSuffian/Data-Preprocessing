```python
import numpy as np
import pandas as pd

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/content/bank_sample.csv')
df.head()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

```python
X = df.iloc[:,:5]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,random_state=30)
```

# Simple Imputer

```python
# First applying simple imputer for comparison
si = SimpleImputer()
X_train_new = si.fit_transform(X_train)
X_test_new = si.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_new)
X_test_scaled = scaler.transform(X_test_new)
```

```python
lr = LogisticRegression()
lr.fit(X_train_scaled,y_train)
y_pred1 = lr.predict(X_test_scaled)
accuracy_score(y_test,y_pred1)
```

# KNN Imputer

```python
knn = KNNImputer(n_neighbors=5,weights='uniform')
X_train_knn = knn.fit_transform(X_train)
X_test_knn = knn.transform(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_knn)
X_test_scaled = scaler.transform(X_test_knn)
```

```python
lr = LogisticRegression()
lr.fit(X_train_scaled,y_train)
y_pred1 = lr.predict(X_test_scaled)
accuracy_score(y_test,y_pred1)
```

# Iterative Imputer (MICE)

```python
from sklearn.impute import IterativeImputer
```

```python
mice = IterativeImputer()
X_train_mice = mice.fit_transform(X_train)
X_test_mice = mice.transform(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_mice)
X_test_scaled = scaler.transform(X_test_mice)
```

```python
lr = LogisticRegression()
lr.fit(X_train_scaled,y_train)
y_pred1 = lr.predict(X_test_scaled)
accuracy_score(y_test,y_pred1)
```
